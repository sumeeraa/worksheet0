```python
#TASK1
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
low_study = []
moderate_study = []
high_study = []

for study, entertainment, sleep in time_data:
    if study < 3:
        low_study.append(study)
    elif 3 <= study <= 5:
        moderate_study.append(study)
    else:
        high_study.append(study)

print("Low Study Hours:", low_study)
print("Moderate Study Hours:", moderate_study)
print("High Study Hours:", high_study)
```

```python
#TASK6
def sum_nested_list(nested_list):
    """
    Recursively calculate the sum of all numbers in a nested list.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print(sum_nested_list(nested_list))  

```

```python
#TASK7
def generate_permutations(s):
    """
    Recursively generate all unique permutations of a string.
    """
    
    if len(s) == 1:
        return [s]
    perms = set()  
    for i in range(len(s)):
        current_char = s[i]
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            perms.add(current_char + perm)

    return list(perms)
print(generate_permutations("abc"))
print(generate_permutations("aab"))
   
```

```python
#TASK9
def min_coins(coins, amount):
    """
    Returns the minimum number of coins needed to make the given amount.
    Uses dynamic programming (tabulation).
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1
print(min_coins([1, 2, 5], 11))   
```

```python
#TASK10
def longest_common_subsequence(s1, s2):
    """
    Computes the length of the Longest Common Subsequence (LCS)
    between two strings using DP (tabulation).
    """
    n, m = len(s1), len(s2)
    dp = [[0] * (m + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = 1 + dp[i - 1][j - 1]
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1]) 
    return dp[n][m]
print(longest_common_subsequence("abcde", "ace"))  
```

```python
#TASK11
def knapsack(weights, values, capacity):
    """
    Solves the 0/1 Knapsack problem using dynamic programming.
    Returns the maximum value that can be achieved.
    """
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        w = weights[i - 1]
        v = values[i - 1]
        for c in range(1, capacity + 1):
            if w <= c: 
                dp[i][c] = max(dp[i - 1][c], v + dp[i - 1][c - w])
            else:
                dp[i][c] = dp[i - 1][c]
    return dp[n][capacity]
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))  # Expected output: 9
```

```python
#TASK8
def calculate_directory_size(directory):
    """
    Recursively calculate the total size of files in a directory structure.
    """
    total_size = 0
    
    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value
    
    return total_size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print(calculate_directory_size(directory_structure))  # Expected: 1400
```
