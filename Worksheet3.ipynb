```python
#Question3
#3.1
result = (
    titanic.assign(fare_per_year = titanic['Fare'] / titanic['Age'])
           .loc[lambda df: df['fare_per_year'] > 5]
           .sort_values(by='fare_per_year', ascending=False)
           [['Name', 'fare_per_year']]
)

print(result)
#3.2
# 1. Total Fare across all passengers
total_fare = titanic['Fare'].sum()

# 2. Total Fare for each class
fare_class1 = titanic[titanic['Pclass'] == 1]['Fare'].sum()
fare_class2 = titanic[titanic['Pclass'] == 2]['Fare'].sum()
fare_class3 = titanic[titanic['Pclass'] == 3]['Fare'].sum()

# 3. Combine totals into a list
fare_totals = [fare_class1, fare_class2, fare_class3]

# 4. Calculate the proportion (percent) of total fare revenue by class
fare_percent = [f / total_fare * 100 for f in fare_totals]

# 5. Display the result
for i, pct in enumerate(fare_percent, start=1):
    print(f"Pclass {i}: {pct:.2f}% of total fare revenue")

#3.3
# 1. Create a new column 'age_group'
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

titanic['age_group'] = titanic['Age'].apply(categorize_age)

# 2. Total number of passengers
total_passengers = len(titanic)

# 3. Count the number of passengers in each age group
age_group_counts = titanic['age_group'].value_counts()

# 4. Calculate the proportion of each age group
age_group_percent = (age_group_counts / total_passengers) * 100

# 5. Display the result
print(age_group_percent)
```
