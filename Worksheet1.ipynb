```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]]);

#1. Prove A.A−1 = I.
A_inverse= np.linalg.inv(A)
dot_product= np.dot(A,A_inverse)
print(dot_product)  #it is approximateto [1 0]
#                                        [0 1]


#2. Prove AB ̸= BA.
AB= np.dot(A, B)
BA= np.dot(B, A)
print(f"\n {AB}")
print(f"\n {BA}")
print("hence, AB is not equal to BA")

#3. Prove (AB)^T = B^T*A^T
AB_T = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)
print("(AB)^T =\n", AB_T)
print("B^T A^T =\n", BT_AT)
print("hence, (AB)^T is not equal to B^T* A^T")


#4 Solve the following system of Linear equation using Inverse Methods.

# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

A_matrix = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])
B_matrix = np.array([[-1],
                  [-3],
                  [9]])

A_inverses= np.linalg.inv(A_matrix)
X= np.dot(A_inverses, B_matrix)
print("the values of x,y,z are: \n",X)
```
